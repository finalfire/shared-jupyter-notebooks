```python
%matplotlib inline
import numpy as np
import os
import pandas as pd
```

```python
def parse(p):
    with open(p) as f_in:
        answer_set, cost = [l.rstrip() for l in f_in][-3:-1]
    
    cliques = list()
    edges = list()
    
    cost = int(cost.split(' ')[-1].split('@')[0])
    for predicate in answer_set[1:-1].split(', '):
        if 'clique' in predicate:
            cliques.append(int(predicate.split('(')[-1].split(')')[0]))
            continue
        
        if 'inFinal' in predicate:
            x,y = list(map(int, predicate.split('(')[-1].split(')')[0].split(',')))
            edges.append((x,y))
            continue
    
    return cost, cliques, edges
```

```python
cols = ('Patient', 'Iteration', 'Stage', 'AS cost', 'Nodes in clique', 'Edges in clique', 'Nodes', 'Edges')
stages = ('CIS', 'RR', 'SP', 'PP')

data = list()
for stage in stages:
    for p in os.listdir(stage):
        if '.ipynb' not in p:
            iteration = int(p[-1])
            p_name = p.split('_')[0]
            
            cost, cliques, edges = parse('{}/{}'.format(stage, p))
            data.append((
                p_name, iteration, stage, cost, len(cliques), len(edges),
                ' '.join(map(str, cliques)), 
                ' '.join(map(lambda x: '{},{}'.format(x[0], x[1]), edges))
            ))
```

```python
df = pd.DataFrame(data, columns=cols)
```
