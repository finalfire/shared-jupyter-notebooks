```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results_brutefor.csv', delimiter=';')
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(2,2, figsize=(24,12))


df[(df['Stage'] == 'CIS')].groupby('N_Edges_Removed').mean().plot(kind='line', ax=ax[0][0], title='AVG OUTCLASS variation, stage CIS')
df[(df['Stage'] == 'PP')].groupby('N_Edges_Removed').mean().plot(kind='line', ax=ax[0][1], title='AVG OUTCLASS variation, stage PP')
df[(df['Stage'] == 'RR')].groupby('N_Edges_Removed').mean().plot(kind='line', ax=ax[1][0], title='AVG OUTCLASS variation, stage RR')
df[(df['Stage'] == 'SP')].groupby('N_Edges_Removed').mean().plot(kind='line', ax=ax[1][1], title='AVG OUTCLASS variation, stage SP');
```
